```python
import pandas as pd

caminhoes_qtc = pd.read_excel("querotruck.xlsx")

caminhoes_qtc["marca"] = (caminhoes_qtc["Modelo"].fillna("").astype(str).str.strip().str.split().str[0])

caminhoes_qtc["versao"] = (caminhoes_qtc["Modelo"].fillna("").astype(str).str.strip().str.split(n=1).str[1])

caminhoes_qtc.to_excel("Caminhoes_QTCK.xlsx")
```
